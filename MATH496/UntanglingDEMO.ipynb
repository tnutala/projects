```python
import numpy as np
import random
import itertools as it
import operator as op


def tangled_graph(hsets,plot=False):
    edgesets=[]
    for h in hsets:
        edgesets.append(it.combinations(h,2))

    for i,j in enumerate(edgesets):
        edgesets[i]=list(j)
        
    g = Graph(multiedges=True)
    for h in edgesets:
        for edges in h:
            g.add_edge(edges)
    if plot==True:
        g.show() 
    return g.incidence_matrix()


def anomalies(mat): #insert incidence matrix here
    n=np.shape(mat)[0]
    E = np.shape(mat)[1]
    anomaly=0    
    part1=[]
    for i in range(n):
        l=list(it.combinations(list(range(n)),i))
        part1.append(l)

    part1=part1[1:ceil(n/2)]  ###TODO: make sure this is indexing right and maybe shorten to reduce runtime
    part1 =[part for parts in part1 for part in parts] ## flattening part1 into one list
#    part1=part1[1:]

    anom=[]
    for part in part1:
        anomaly=0
        spec=mat[part,:]
        for col in range(E):
            if list(sum(spec[:,col]))!=[1]:
                anomaly+=1
            else:
                continue
        anom.append(anomaly)
    
    return anom,part1
    
def hset_rand(ksize,snum,days):
    hsets=[]

    if ksize == 'random':
        for i in range(np.random.randint(snum,snum+1)):    #number of schedules (inclusive,exclusive)
            hset= random.sample(range(days), np.random.randint(2,days+1)) #generates schedules; randint(smallest allowed clique, largest allowed)
            hsets.append(hset)
        
    else:
        for i in range(np.random.randint(snum,snum+1)):    #number of schedules (inclusive,exclusive)
            hset= random.sample(range(days), np.random.randint(ksize,ksize+1)) #generates schedules; randint(smallest allowed clique, largest allowed)
            hsets.append(hset)
    return hsets


def ncr(n,r):
    if n<r:
        return 0
    elif n==r or r==0:
        return 1
    else:
        r = min(r, n-r)
        numer = reduce(op.mul, xrange(n, n-r, -1), 1)
        denom = reduce(op.mul, xrange(1, r+1), 1)
        return numer//denom
```
